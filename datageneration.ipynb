```python
import pandas as pd
import random
from faker import Faker
from datetime import date, timedelta

fake = Faker()
random.seed(42)

# Kenyan first names (male and female)
kenyan_first_names_male = [
    "Brian", "Kevin", "Dennis", "Eric", "Samuel", "Joseph", "Daniel", "David",
    "James", "John", "Peter", "Paul", "Michael", "George", "Patrick", "Francis",
    "Alex", "Victor", "Collins", "Edwin", "Leonard", "Martin", "Bernard", "Raymond",
    "Geoffrey", "Kelvin", "Clinton", "Elijah", "Moses", "Isaac", "Emmanuel", "Felix"
]

kenyan_first_names_female = [
    "Faith", "Grace", "Mercy", "Joy", "Winnie", "Aisha", "Amina", "Sharon",
    "Caroline", "Esther", "Beatrice", "Doris", "Lilian", "Vivian", "Irene",
    "Tabitha", "Judith", "Agnes", "Pauline", "Stella", "Naomi", "Lydia",
    "Rehema", "Fatuma", "Zawadi", "Wanjiku", "Njeri", "Wangari", "Mwende", "Nduta"
]

# Kenyan middle names (often Christian or Swahili)
kenyan_middle_names = [
    "Wanjiku", "Kamau", "Otieno", "Mwangi", "Odhiambo", "Njoroge", "Kariuki",
    "Gitau", "Kiprotich", "Chebet", "Akinyi", "Adhiambo", "Wambui", "Ndungu",
    "Mutua", "Musyoka", "Muthoni", "Wairimu", "Nyambura", "Waweru", "Gichuru",
    "Kimani", "Njenga", "Mugo", "Kinuthia", "Kinyanjui", "Wekesa", "Simiyu",
    "Barasa", "Juma", "Hassan", "Abdi", "Omar", "Ali", "Mohamed", "Omondi"
]

# Kenyan surnames (from various communities)
kenyan_surnames = [
    "Kamau", "Otieno", "Mwangi", "Odhiambo", "Njoroge", "Kariuki", "Gitau",
    "Kiprotich", "Cheruiyot", "Akinyi", "Mutua", "Musyoka", "Waweru", "Gichuru",
    "Kimani", "Njenga", "Mugo", "Kinuthia", "Wekesa", "Simiyu", "Barasa",
    "Juma", "Hassan", "Omondi", "Ogolla", "Odundo", "Owino", "Adhiambo",
    "Muthoni", "Nyambura", "Ndungu", "Muthee", "Gitahi", "Ngugi", "Mbugua",
    "Macharia", "Kuria", "Gacheru", "Muriuki", "Njeru", "Kitheka", "Musau",
    "Kivuva", "Mwenda", "Muthama", "Makau", "Mutuku", "Kyalo", "Mbatha",
    "Limo", "Ruto", "Koech", "Bett", "Kiptoo", "Lagat", "Ngetich", "Koskei"
]

# Departments and positions
department_positions = {
    "Engineering": ["Software Engineer", "Senior Software Engineer", "QA Engineer", "DevOps Engineer", "Engineering Manager"],
    "Finance": ["Accountant", "Senior Accountant", "Finance Analyst", "Finance Manager", "CFO"],
    "Human Resources": ["HR Officer", "HR Specialist", "Recruitment Officer", "HR Manager", "Head of HR"],
    "Sales": ["Sales Representative", "Senior Sales Rep", "Sales Executive", "Sales Manager", "Head of Sales"],
    "Operations": ["Operations Officer", "Logistics Coordinator", "Operations Analyst", "Operations Manager", "COO"],
    "Marketing": ["Marketing Officer", "Content Creator", "Digital Marketer", "Marketing Manager", "Head of Marketing"],
    "ICT": ["IT Support", "Network Engineer", "Systems Administrator", "IT Manager", "CTO"],
    "Legal": ["Legal Officer", "Compliance Officer", "Legal Counsel", "Senior Counsel", "Head of Legal"],
    "Customer Service": ["Customer Rep", "Senior Customer Rep", "Team Lead", "Customer Service Manager", "Head of CS"],
    "Procurement": ["Procurement Officer", "Senior Procurement Officer", "Procurement Manager", "Head of Procurement"]
}

# Salary ranges per position seniority (KES)
salary_ranges = {
    0: (35000, 55000),    # Junior
    1: (55000, 85000),    # Mid
    2: (50000, 75000),    # Specialist
    3: (90000, 140000),   # Manager
    4: (150000, 300000),  # Head/C-suite
}

# Kenyan banks
kenyan_banks = [
    "KCB Bank", "Equity Bank", "Co-operative Bank", "NCBA Bank",
    "Absa Bank Kenya", "Standard Chartered Kenya", "I&M Bank",
    "Family Bank", "DTB Bank", "Stanbic Bank Kenya"
]

print(" Setup complete — Kenyan data pools loaded")
```

```python
def generate_kenyan_name():
    gender = random.choice(["Male", "Female"])
    if gender == "Male":
        first = random.choice(kenyan_first_names_male)
    else:
        first = random.choice(kenyan_first_names_female)
    middle = random.choice(kenyan_middle_names)
    last = random.choice(kenyan_surnames)
    return first, middle, last, gender

def generate_national_id():
    # Kenya National ID: 7-8 digit number
    return str(random.randint(10000000, 40000000))

def generate_sha_number():
    # SHA (Social Health Authority) number format: SHA + 10 digits
    return "SHA" + str(random.randint(1000000000, 9999999999))

def generate_nssf_number():
    # NSSF Kenya format: 7 digit number
    return str(random.randint(1000000, 9999999))

def generate_bank_details(name):
    bank = random.choice(kenyan_banks)
    # Kenyan bank account numbers are typically 13 digits
    account_number = str(random.randint(1000000000000, 9999999999999))
    return bank, account_number

def generate_kra_pin(name_first, national_id):
    # KRA PIN format: A + 9 digits + letter
    letters = "ABCDEFGHJKLMNPQRSTUVWXYZ"
    return "A" + national_id[:8].zfill(8) + random.choice(letters)

def generate_salary(position_index):
    low, high = salary_ranges[position_index]
    base = random.randint(low // 1000, high // 1000) * 1000
    return base

def generate_bonus(salary, position_index):
    # Bonus as % of salary — higher positions get higher bonus potential
    bonus_pct = [0.05, 0.08, 0.07, 0.12, 0.20][position_index]
    bonus = round(salary * random.uniform(0, bonus_pct) / 1000) * 1000
    return bonus

def generate_hire_date():
    start = date(2015, 1, 1)
    end = date(2024, 12, 31)
    delta = (end - start).days
    return start + timedelta(days=random.randint(0, delta))

def generate_leave_days(hire_date):
    # Annual leave entitlement in Kenya is 21 days
    # Remaining depends on how much has been used
    years_served = (date.today() - hire_date).days / 365
    total_entitlement = min(int(years_served) * 21, 63)  # cap at 3 years accrual
    used = random.randint(0, min(total_entitlement, 21))
    return max(0, total_entitlement - used)

print(" Generator functions ready")
```

```python
employees = []

for emp_id in range(1, 301):
    first, middle, last, gender = generate_kenyan_name()
    full_name = f"{first} {middle} {last}"
    title = "Mr." if gender == "Male" else random.choice(["Ms.", "Mrs."])
    full_name = f"{first} {middle} {last}"
    
    department = random.choice(list(department_positions.keys()))
    positions = department_positions[department]
    position_index = random.choices([0,1,2,3,4], weights=[35,30,15,15,5])[0]
    position = positions[min(position_index, len(positions)-1)]
    
    national_id = generate_national_id()
    sha = generate_sha_number()
    nssf = generate_nssf_number()
    kra_pin = generate_kra_pin(first, national_id)
    bank, account_no = generate_bank_details(full_name)
    
    salary = generate_salary(position_index)
    bonus = generate_bonus(salary, position_index)
    
    # PAYE and NSSF deductions (Kenya)
    nssf_deduction = min(2160, salary * 0.06)   # NSSF: 6% capped at KES 2,160
    gross = salary + bonus
    taxable = gross - nssf_deduction
    # Simplified PAYE bands
    if taxable <= 24000:
        paye = taxable * 0.10
    elif taxable <= 32333:
        paye = 2400 + (taxable - 24000) * 0.25
    elif taxable <= 500000:
        paye = 4483 + (taxable - 32333) * 0.30
    else:
        paye = 144483 + (taxable - 500000) * 0.35
    paye = max(0, paye - 2400)  # personal relief
    
    net_salary = gross - nssf_deduction - paye
    
    hire_date = generate_hire_date()
    leave_days = generate_leave_days(hire_date)
    
    dob = date(
        random.randint(1975, 2000),
        random.randint(1, 12),
        random.randint(1, 28)
    )
    
    phone = "07" + str(random.randint(10000000, 99999999))
    email = f"{first.lower()}.{last.lower()}{random.randint(1,99)}@company.co.ke"
    
    employees.append({
        "Employee ID":      f"EMP-{emp_id:03d}",
        "Full Name":        full_name,
        "First Name":       first,
        "Middle Name":      middle,
        "Last Name":        last,
        "Gender":           gender.capitalize(),
        "Date of Birth":    dob.strftime("%Y-%m-%d"),
        "National ID":      national_id,
        "KRA PIN":          kra_pin,
        "SHA Number":       sha,
        "NSSF Number":      nssf,
        "Phone Number":     phone,
        "Email":            email,
        "Department":       department,
        "Position":         position,
        "Seniority Level":  ["Junior","Mid-level","Specialist","Manager","Senior Leader"][position_index],
        "Hire Date":        hire_date.strftime("%Y-%m-%d"),
        "Basic Salary (KES)":    salary,
        "Bonus (KES)":           bonus,
        "Gross Salary (KES)":    gross,
        "NSSF Deduction (KES)":  round(nssf_deduction),
        "PAYE (KES)":            round(paye),
        "Net Salary (KES)":      round(net_salary),
        "Bank Name":             bank,
        "Bank Account No":       account_no,
        "Leave Days Remaining":  leave_days,
        "Employment Status":     random.choices(["Active","Active","Active","Active","On Leave","Probation"], weights=[70,70,70,70,10,10])[0],
        "Title":    title,
        "Gender":   gender,
    })

df = pd.DataFrame(employees)
print(f" Generated {len(df)} employee records")
print(f"\nDepartment breakdown:")
print(df["Department"].value_counts().to_string())
print(f"\nSeniority breakdown:")
print(df["Seniority Level"].value_counts().to_string())
print(df["Gender"].value_counts())
```
